```python
from collections import namedtuple 
from itertools import product
from itertools import combinations
```

```python
presidente = ["A","B","C","D","E"]
```

```python
Chapa = namedtuple("Chapa", "Presidente VicePresidente")
Chapas = [ Chapa(*x) for x in itertools.permutations(presidente,2)]
```
